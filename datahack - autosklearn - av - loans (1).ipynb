```python
import multiprocessing
import shutil

import sklearn.model_selection
import sklearn.datasets
import sklearn.metrics

import autosklearn
from autosklearn.metrics import accuracy
from autosklearn.classification import AutoSklearnClassifier
from autosklearn.constants import *

automl = autosklearn.classification.AutoSklearnClassifier(
        time_left_for_this_task=120,
        per_run_time_limit=30,
        tmp_folder='/home/sunil/Desktop/datahack/autosklearn-tmp1',
        output_folder ='/home/sunil/Desktop/datahack/autosklearn-op1',
        ensemble_size=1,
        delete_tmp_folder_after_terminate=False,
        resampling_strategy='cv',
        resampling_strategy_arguments={'folds': 5},)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
automl.fit(X_train.copy(), y_train.copy())
```

```python
automl.cv_results_
```

```python
automl.refit(X_train.copy(), y_train.copy())
```

```python
print(automl.show_models())
```

```python
predictions = automl.predict(X_test)
```

```python
predictions
```

```python
print("Accuracy score", sklearn.metrics.accuracy_score(y_test, predictions))
```

```python
av_test = pd.read_csv('processed_test.csv')
temp=av_test['Loan_ID']
del av_test['Loan_ID']
av_test_array=av_test.values
results = automl.predict(av_test_array)
results=pd.Series(results)
results.replace([0, 1], ['N','Y'],inplace=True)
final=pd.concat([temp, results], axis=1)
final.rename(columns={0: 'Loan_Status'}, inplace=True)
file_name="autosklearn_final_av_test.csv"
final.to_csv(file_name, sep=',', encoding='utf-8',index=False)
```
