```python
X_train_h2o.describe()
```

```python
features = X_train.columns.values.tolist()
target = "class"
```

```python
auto_h2o = H2OAutoML(max_runtime_secs=60*3)
auto_h2o.train(x=features,
               y=target,
               training_frame=X_train_h2o)
```
